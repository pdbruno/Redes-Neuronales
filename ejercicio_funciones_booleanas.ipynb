# Funciones Booleanas

Implementar un perceptrón simple para tratar de aprender los siguientes
conjuntos de datos para dos variables.
1. Las funciones lógicas AND y OR simultáneamente.
2. Las función lógica XOR.

Entrenarlo probando todas las combinaciones de las siguientes posibilidades:

1. Con función de activación y valores de datos binarios y bipolares.
2. Con función de activación escalón y sigmoidea.
3. Con y sin unidad de umbral

```python
# Empecemos por importar la librería numpy que nos va a ayudar con las operaciones algebráicas.
import numpy as np
from matplotlib import pyplot as mpl, cm
from perceptron_simple import PerceptronSimple 
from testing_utils import posibilidadesUmbral, posibilidadesFuncionActivacion, posibilidadesTipoDato
```

```python
# Y establezcamos las dimensiones de los datos que vamos a utilizar.

N = 2
M = 2
P = 4
```

```python
logic_input = np.array([[0, 0],
                        [0, 1],
                        [1, 0],
                        [1, 1]])

logic_target = np.array([[0, 0],
                         [0, 1],
                         [0, 1],
                         [1, 1]])
```

```python
tipoDato = posibilidadesTipoDato[1]
funcActivacion = posibilidadesFuncionActivacion[0]
umbral = posibilidadesUmbral[1]
pc = PerceptronSimple(umbral.getN(N), M, funcActivacion.getG(tipoDato.getTipoDato()), 1e-4)
E = pc.fit(umbral.getX(tipoDato.getX(logic_input)), tipoDato.getZ(logic_target))
print(
    f'{tipoDato.getName()}\n{funcActivacion.getName()}\n{umbral.getName()}\nEl último error es de {E[-1]}')
mpl.plot(E, 'r')
mpl.show()
```

# Arrancamos con lo que se dice testear!!!

```python
for tipoDato in posibilidadesTipoDato:
    for funcActivacion in posibilidadesFuncionActivacion:
        for umbral in posibilidadesUmbral:
            pc = PerceptronSimple(umbral.getN(N), M, funcActivacion.getG(tipoDato.getTipoDato()), 1e-4)
            E = pc.fit(umbral.getX(tipoDato.getX(logic_input)), tipoDato.getZ(logic_target))
            print(
                f'{tipoDato.getName()}\n{funcActivacion.getName()}\n{umbral.getName()}\nEl último error es de {E[-1]}')
            mpl.plot(E, 'r')
            mpl.show()
```
