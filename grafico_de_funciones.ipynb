```python
import numpy as np
from matplotlib import pyplot as mpl
 
```

```python
dominio = np.linspace(-2, 2, 100)
print(dominio.shape)
```

```python
f1 = lambda x: 1/(1 + np.exp(-x))
f2 = np.tanh
f3 = np.sign
f4 = lambda x: f1(x)*(1-f1(x))
f5 = lambda x: 1 - f2(x)**2
```

```python
y1 = f1(dominio)
y2 = f2(dominio)
y3 = f3(dominio)
y4 = f4(dominio)
y5 = f5(dominio)
```

```python
g = mpl.plot( dominio, y1)
mpl.show()
g = mpl.plot( dominio, y2)
mpl.show()
g = mpl.plot( dominio, y3)
mpl.show()
g = mpl.plot( dominio, y4)
mpl.show()
g = mpl.plot( dominio, y5)
mpl.show()
```
